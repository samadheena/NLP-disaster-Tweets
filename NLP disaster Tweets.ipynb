```python
import spacy
# Load the spaCy English model
nlp = spacy.load('en_core_web_sm')
```

```python
# Define a function named preprocess which takes a text input
def preprocess(text):
    # Convert the text to lowercase
    text = text.lower()
    
    # Assuming wtk is a function that tokenizes the text, splitting it into individual words
    text = wtk(text)
    
    # Initialize an empty list to store the tokenized words after filtering out non-alphanumeric characters
    txt = []
    
    # Iterate through each tokenized word in the text
    for i in text:
        # If the word consists of alphanumeric characters (letters or numbers), append it to the txt list
        if i.isalnum():
            txt.append(i)
    
    # Initialize an empty list to store the final list of words after removing stopwords
    words = []
    
    # Iterate through each tokenized word in txt
    for i in txt:
        # If the word is not a stopword in the English language, append it to the words list
        if i not in stopwords.words('english'):
            words.append(i)
    
    # Concatenate the words in the words list into a single string separated by whitespace
    words = " ".join(words)
    
    # Assuming nlp is an instance of a natural language processing (NLP) model (e.g., SpaCy)
    # Process the text to perform lemmatization
    word = nlp(words)
    
    # Initialize an empty list to store the lemmatized forms of the words
    lemma = []
    
    # Iterate through each tokenized word processed by the NLP model
    for i in word:
        # Append its lemmatized form to the lemma list
        lemma.append(i.lemma_)
    
    # Return the list of lemmatized words
    return lemma
```

```python
list=['deed','http','maker','woolen']
list
```

```python
def remove_http(list):
    # Initialize an empty list to store the filtered words
    new_list = []
    
    # Iterate through each word in the input list
    for word in list:
        # Check if the word is not equal to 'http'
        if word != 'http':
            # If the word is not 'http', append it to the new_list
            new_list.append(word)
    
    # Return the filtered list without 'http' occurrences
    return new_list
```
